Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

YA dataset called `schools.csv` is provided, which is previewed below.

This work answers three key questions about New York City (NYC) public school SAT performance:
- Which schools are the best math schools in NYC?
- Which are the top 10 schools in NYC based on the overall SAT scores (math, reading, writing)?
- What school has the greatest standard deviation in terms of the overall SAT scores of students?

## Setting up data

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

## Creating dataframe: best_math_schools

```python
# Schools with at least 80% in Math SAT: best_math_schools

max_score = 800
score_80 = 0.8 * max_score
best_math_schools = schools[schools['average_math'] >= score_80][['school_name','average_math']].sort_values('average_math',ascending = False)

best_math_schools
```

```python
# Top 10 best schools based on overall SAT scores (math, reading, and writing: top_10_schools

schools['total_SAT'] = schools.average_math + schools.average_reading + schools.average_writing
top_10_schools = schools.groupby("school_name")["total_SAT"].mean().reset_index().sort_values("total_SAT", ascending=False).head(10)

top_10_schools
```

```python
# Calculating for schools with largest std_dev per NYC borough: largest_std_dev

summary = schools.groupby('borough')['total_SAT'].agg(['count','mean','std']).round(2)
summary.columns = ['num_schools','average_SAT','std_SAT'] # renaming column names

largest_std_dev = summary[summary['std_SAT'] == max(summary['std_SAT'])]

largest_std_dev
```
